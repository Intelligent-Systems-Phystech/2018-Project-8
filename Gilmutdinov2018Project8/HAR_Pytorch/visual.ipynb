```python
trainX, trainy, testX, testy = load_dataset()
trainX, testX = scale_data(trainX, testX, standartize = True)
```

```python
LABELS = [
    "WALKING",
    "WALKING_UPSTAIRS",
    "WALKING_DOWNSTAIRS",
    "SITTING",
    "STANDING",
    "LAYING"
]
```

```python
class MyDataset(Dataset):

    def __init__(self, data, labels, transform = None):
        self.transform = transform
        self.labels = labels.flatten()
        if self.transform is not None:
            self.data = data

    def __getitem__(self, item):
        label, row = self.labels[item], self.data[item]
        if self.transform is not None:
            row = self.transform(row)
        return row, label

    def __len__(self):
        return len(self.labels)

def toTensor(data):
    return torch.from_numpy(data.transpose((1, 0))).float()
    

transform = transforms.Compose([toTensor])
```

```python
BATCH_SIZE = 2947
Test_data = MyDataset(testX, testy, transform=transform)
testloader = DataLoader(dataset=Test_data,
                              batch_size = BATCH_SIZE,
                              shuffle=False,
                              )
```

```python
use_cuda = torch.cuda.is_available()
```

```python
with h5py.File('acc_loss.hdf5', 'r') as f:
    Accuracy = f['accuracy']
    Loss = f['loss']
    Accuracy = Accuracy[:]
    Loss = Loss[:]
```

```python
def plot_func(Accuracy, Loss, xlabel='Epoch', figsize=(20, 20)):
    
    f, ax = plt.subplots(2, 1, figsize=figsize)
    epochs = np.arange(0, Loss.shape[0])
    ax[0].plot(epochs, Accuracy, '-', ms=10)
    ax[1].plot(epochs, Loss)
    
    ax[0].set_title('Accuracy depends on epoch')
    ax[0].set_xlabel(xlabel)
    ax[0].set_ylabel('Accuracy score')
    
    ax[1].set_title('Loss depends on epoch')
    ax[1].set_xlabel(xlabel)
    ax[1].set_ylabel('Loss per epoch')
    
    plt.ticklabel_format(axis='y',style='sci',scilimits=(1,3))
    plt.show()
```

```python
plot_func(Accuracy, Loss)
```

```python
model = Model()
model.load_state_dict(torch.load('saved_model_state.pt'))
```

```python
model.eval()
correct = 0
total = 0
# Iterate through test dataset

with torch.no_grad():
    dataiter = iter(testloader)
    inputs, labels = dataiter.next()
    outputs = model(inputs)
    _, predicted = torch.max(outputs.data, 1)
    total = labels.size(0)
    correct = (predicted == labels).sum().item()
print('Accuracy : %.3f' %(correct/total) )
print("Precision: %.3f" %(metrics.precision_score(labels, predicted, average="weighted")))
print("Recall: %.3f" %(metrics.recall_score(labels, predicted, average="weighted")))
print("f1_score: %.3f" %(metrics.f1_score(labels, predicted, average="weighted")))
```

```python
print("")
print("Confusion Matrix:")
confusion_matrix = metrics.confusion_matrix(labels, predicted)
print(confusion_matrix)
```
